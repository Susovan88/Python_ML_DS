```
batter=pd.read_csv('/content/batter.csv')
batter.head()
```

```
batter=batter.sample(30)

plt.figure(figsize=(12,7))
plt.scatter(batter['avg'],batter['strike_rate'],s=batter['runs'])

## horizontal and vertical lines
plt.axhline(140,color='red')
plt.axhline(130,color='pink')
plt.axvline(40,color='green')
plt.axvline(30,color='yellow')

## each point label
for i in range(batter.shape[0]):
  plt.text(batter['avg'].values[i],batter['strike_rate'].values[i],batter['batter'].values[i])

plt.show()
```

```
### surface plot 3D

x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
```

```
xx,yy=np.meshgrid(x,y)
z=xx**2+yy**2

fig=plt.figure()
ax=plt.subplot(projection='3d',)
p=ax.plot_surface(xx,yy,z,cmap='winter')

fig.colorbar(p)
```

```
z=np.sin(xx)+np.cos(yy)

fig=plt.figure()
ax=plt.subplot(projection='3d')

p1=ax.plot_surface(xx,yy,z,cmap='viridis')

fig.colorbar(p1)
```

```
z=np.sin(xx)+np.log(yy)

fig=plt.figure()
ax=plt.subplot(projection='3d')

p1=ax.plot_surface(xx,yy,z,cmap='viridis')

fig.colorbar(p1)
```

## Contour Plot -> 3D graph represent into 2D

```
xx,yy=np.meshgrid(x,y)
z=xx**2+yy**2

fig=plt.figure()
ax=plt.subplot()
p=ax.contour(xx,yy,z,cmap='winter')

fig.colorbar(p)
```

```
xx,yy=np.meshgrid(x,y)
z=xx**2+yy**2

fig=plt.figure()
ax=plt.subplot()
p=ax.contourf(xx,yy,z,cmap='winter')

fig.colorbar(p)
```

```
z=np.sin(xx)+np.cos(yy)

fig=plt.figure()
ax=plt.subplot()

p1=ax.contourf(xx,yy,z,cmap='viridis')

fig.colorbar(p1)
```

## Heatmap

```
balls=pd.read_csv('/content/IPL_Ball_by_Ball_2008_2022.csv')
delivary=balls[(balls['ballnumber'].isin([1,2,3,4,5,6])) & (balls['batsman_run']==6)]
```

```
grid=delivary.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')

fig=plt.figure()
plt.imshow(grid)
plt.yticks(delivary['overs'].unique(),list(delivary['overs'].unique()+1))
plt.xticks(list(np.arange(0,6)),list(np.arange(1,7)))
plt.colorbar()
plt.show()
```
