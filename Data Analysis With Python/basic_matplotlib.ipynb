```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

## 2D line plot 
- Bivariate Analysis
- categorical -> numerical and numerical -> numerical
- Use case - Time series data

```python
data= pd.read_csv("sharma-kohli.csv")
data
```

```python
plt.plot(data['index'],data["V Kohli"],color="green",linestyle="solid",linewidth=3)
plt.plot(data['index'],data["RG Sharma"],color="red",linestyle="solid",linewidth=3)
plt.title("VK vs RS")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
plt.plot(data['index'],data["V Kohli"],color="green",linestyle="solid",linewidth=3,marker='D')
plt.plot(data['index'],data["RG Sharma"],color="red",linestyle="dashdot",linewidth=3,marker="D")
plt.title("VK vs RS")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
plt.plot(data['index'],data["V Kohli"],color="green",linestyle="solid",linewidth=3,marker='D',label="Virat")
plt.plot(data['index'],data["RG Sharma"],color="red",linestyle="dashdot",linewidth=3,marker="D",label="Rohit")
plt.title("VK vs RS")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
plt.legend(loc='lower right')
plt.grid()
plt.show()
```

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.xlim(2016,2020)
plt.ylim(40000,70000)
plt.grid()
```

## Bar chart
- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color="red")
```

```python
plt.barh(colors,children)
```

```python
data=pd.read_csv("batsman_season_record.csv")

plt.bar(np.arange(data.shape[0])-0.2,data['2015'],width=0.2)
plt.bar(np.arange(data.shape[0]),data['2016'],width=0.2)
plt.bar(np.arange(data.shape[0])+0.2,data['2017'],width=0.2)
plt.xticks(np.arange(data.shape[0]),data['batsman'])
plt.xticks(rotation='vertical')
plt.grid()

plt.show()
```

```python
plt.bar(data['batsman'],data['2015'],label='2015')
plt.bar(data['batsman'],data['2016'],bottom=data['2015'],label='2016')
plt.bar(data['batsman'],data['2017'],bottom=(data['2015']+data['2016']),label='2017')
plt.legend()
```
