Resampling in Pandas (Simple to Advanced)

Resampling is used to change the frequency of time series data (either increasing or decreasing it).

1. Downsampling (Reduce Frequency)

Example: Convert daily data to monthly data.

2. Upsampling (Increase Frequency)

Example: Convert monthly data to daily data.

```python
google['Close'].resample('M').mean()   ## downsampling
```

```python
## downsampling
google['Close'].plot()
google['Close'].resample('M').mean().plot()
```

```python
google['Close'].plot()
google['Close'].resample('Y').mean().plot()
```

```python
#### up sampling
google['Close'].resample('H').interpolate(method='linear')
```

```python
google['Close'].resample('H').interpolate(method='linear').plot()
```

Rolling Window in Pandas

A rolling window applies a function (like mean, sum, std, etc.) over a sliding window of fixed size across your data.

```python
google[['Close','High','Low','Open']].rolling(5).mean()
```

```python
google[['High','Low']].rolling(500).mean().plot()
```

```python
google['Close'].plot()
google['Close'].rolling(90).mean().plot()
```

Exponential Weighted Moving (EWM) in Pandas

ewm() applies an exponentially weighted function where recent data points get more weight than older ones, unlike rolling() which treats all points equally in a window.

```python
google['Close'].ewm(30).mean().plot()
```

```python
google['Close'].plot()
google['Close'].rolling(90).mean().plot(color='red')
google['Close'].ewm(90).mean().plot(color='black')
```

Shifting in Pandas

shift() moves the values of a column or Series up or down by a specified number of periods while keeping the index fixed. It is often used for lag/lead features in time series.

```python
google['Close'].shift(2)
```

```python
df=pd.read_csv('/content/login.csv',header=None)[[1,2]].rename(columns={1:'user_id',2:'login_time'})
df
```

```python
df['login_time']=pd.to_datetime(df['login_time'])
df['user_id'].value_counts().head(3)
```

```python
## top 3 user avg loging gap
top_user=df['user_id'].value_counts().head(3).index
for i in top_user:
  user_df=df[df['user_id']==i]
  user_df['shifting_Time']=user_df['login_time'].shift(1)
  avg_gap=(user_df['login_time']-user_df['shifting_Time']).mean()
  print(f"user_id:{i} has has avg login gap : {avg_gap}")
```
