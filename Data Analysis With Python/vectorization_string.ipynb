```python
# Load the Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
data = pd.read_csv(url)  ## sample data
data.head()
```

```python
data.info()
```

```python
data["lastName"]=data["Name"].str.strip().str.split(',').str.get(0)
data.head()
```

```python
data[["title","firstName"]]=data["Name"].str.strip().str.split(',').str.get(1).str.strip().str.split(' ',n=1,expand=True)   ## expand =true -> convert into datafream
data.head()
```

```python
data["title"].value_counts()
```

```python
data['tile']=data['title'].str.replace('Ms.','Miss')
data['tile']=data['title'].str.replace('Mlle.','Miss')
```
