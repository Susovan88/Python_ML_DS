## pd.concat() stitches those DataFrames together vertically (row-wise).

On the first loop, it just becomes the filtered data.

On later loops, it stacks the new group’s filtered rows below the existing ones.

```python
### loop in groups    ## from group we get two type of data -> datafrem(data) and str
df=pd.DataFrame(columns=movies.columns)
for group,data in genre:
    df=pd.concat([df,data[data["IMDB_Rating"]==data["IMDB_Rating"].max()]])

df
```
