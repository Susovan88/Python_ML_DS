```python
import matplotlib.pyplot as plt

x=[23,45,60,76,90]
y=[10,34,26,90,24]

plt.plot(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Aixs")
plt.title("Basic line Plot")
plt.show()
```

```python
plt.plot(x,y,color="green",linestyle="--",marker='o',linewidth=1,markersize=4)
# plt.grid(True)
plt.show()
```

```python
x = [1, 12, 3, 4, 15, 6, 7]
y1 = [2, 14, 6, 8, 19, 12, 14]
y2 = [11, 3, 5, 17, 9, 1, 13]
plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(x,y1,color="Blue",linestyle="--",marker='o',linewidth=1,markersize=4)
plt.title("plot 1")

plt.subplot(2,2,2)
plt.plot(x,y2,color="Green",linestyle="--",marker='o',linewidth=1,markersize=4)
plt.title("plot 2")

plt.subplot(2,2,3)
plt.plot(y1,y2,color="Green",linestyle="--",marker='o',linewidth=1,markersize=4)
plt.title("plot 3")

z=[i*2 for i in x]
print(z)

plt.subplot(2,2,4)
plt.plot(x,z,color="Red",linestyle="--",marker='o',linewidth=1,markersize=4)
plt.title("plot 4")
```

```python
### Bar Plor
categories = ["A", "B", "C", "D", "E", "F", "G", "H"]
values = [10, 15, 7, 20, 5, 12, 18, 9]

plt.bar(categories,values,color="Blue")
plt.ylabel("values")
plt.xlabel("categories")
plt.title("Basic Bar Plot")
```

### A Histogram is used to show the distribution of data (how data is spread out).

It divides your data into bins (intervals) and then counts how many data points fall into each bin.

Think of it like:

Grouping similar values together

Showing how many times values appear in each group

###  A bin is like a container (or range) that groups your data.

When you create a histogram:

The data gets split into intervals (bins).

Each bin counts how many data points fall inside it.

Then it draws a bar for each bin ðŸ“Š

```python
## histograms 
data=[10, 20, 22, 35, 40, 45, 55, 60, 75, 80, 85, 90, 95]
plt.hist(data,bins=5,color="gray",edgecolor="black")
```

```python
### create a scatter Plot
x = [1, 12, 3, 12, 3, 4, 15, 6, 7, 4, 15, 6, 7]
y1 = [2, 14, 6, 8, 14, 6, 8, 19, 12, 14, 19, 12, 14]

plt.scatter(x,y1,color="red",marker="*")
```

```python

### pipe chart
labels = ["A", "B", "C", "D", "E", "F", "G", "H"]
values = [10, 23, 13, 18, 5, 7, 16, 8]
colors = [
    "gold",         # ðŸŸ¡ gold
    "lightcoral",   # ðŸ”´ soft coral
    "lightskyblue", # ðŸ”µ light blue
    "palegreen",    # ðŸŸ¢ soft green
    "plum",         # ðŸŸ£ light purple
    "salmon",       # ðŸŸ¥ salmon pink
    "orange",       # ðŸŸ  orange
    "violet"        # ðŸŸª violet
]
explode=(0,0.2,0,0,0,0,0,0)

plt.pie(values,colors=colors,labels=labels,explode=explode,shadow=True,)
```

### What is .reset_index()?
When you use .reset_index(), it resets the index of your DataFrame back to the default numbers 0,1,2,....

```python
sales_on_date=sales_data_df.groupby("Date")["Total Revenue"].sum().reset_index()
plt.plot(sales_on_date["Date"],sales_on_date["Total Revenue"])
```
