```python
def checkBWtwoTeams(team1,team2):
    total_match=ipl[(ipl["Team1"]==team1) & (ipl["Team2"]==team2) | (ipl["Team1"]==team2) & (ipl["Team2"]==team1)]
    team1_wins=total_match[total_match["WinningTeam"]==team1].shape[0]
    team2_wins=total_match[total_match["WinningTeam"]==team2].shape[0]
    return [team1_wins,team2_wins]

checkBWtwoTeams("Kolkata Knight Riders","Chennai Super Kings")
```

### apply() -> series & dataframe

axis=1 -> rows

axis=2 -> cols

```python
points_df = pd.DataFrame(
    {
        '1st_point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd_point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)
```

```python
def euclidean(row):
    point_a=row['1st_point']
    point_b=row['2nd_point']
    return ((point_a[0]-point_b[0])**2 + (point_a[1]-point_b[1])**2)**0.5


points_df['euclidean']=points_df.apply(euclidean,axis=1) 
points_df
```
