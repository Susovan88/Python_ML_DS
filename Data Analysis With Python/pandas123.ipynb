```python

contidion1=movies['genres'].str.split("|").apply(lambda x: "Action" in x)
contidion2=movies["imdb_rating"]>7.5
movies[contidion1 & contidion2].shape[0]
```

```python
team1="Kolkata Knight Riders"
team2="Chennai Super Kings"
```

```python
def checkBWtwoTeams(team1,team2):
    total_match=ipl[(ipl["Team1"]==team1) & (ipl["Team2"]==team2) | (ipl["Team1"]==team2) & (ipl["Team2"]==team1)]
    team1_wins=total_match[total_match["WinningTeam"]==team1].shape[0]
    team2_wins=total_match[total_match["WinningTeam"]==team2].shape[0]
    return [team1_wins,team2_wins]

checkBWtwoTeams("Kolkata Knight Riders","Chennai Super Kings")
```
