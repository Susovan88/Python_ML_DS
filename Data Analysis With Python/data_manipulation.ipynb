#### df.isnull()
Returns a DataFrame of the same shape, with True where the value is NaN (or None), and False
### df.isnull().sum()
First df.isnull() creates that boolean mask; then .sum()—by default along axis=0 (i.e. down each column)—adds up True values (True is treated as 1). This gives you the count of missing values per column
####  df.isnull().any(axis=1)
Again you start with the boolean mask from df.isnull(), but .any(axis=1) collapses each row into a single True/False:
True if any value in that row is missing
False if none are missing
### df.fillna(value)
→ Replace missing values with value (e.g. the mean, zero, or a placeholder).

###  2. df["Value_new"] = df["Value"].fillna(df["Value"].mean()).astype(int)
This does three things in a single line:

# df["Value"].fillna(df["Value"].mean())
Replace missing values (NaN) in the Value column with the column’s mean.

# .astype(int)
Convert the result to integers (since mean will produce floats).

# df["Value_new"] = ...
Store the result in a new column called Value_new.

```python
df["Value_new"]=df["Value"].fillna(df["Value"].mean()).astype(int)
df.head()
```

### What does df["Value"].apply(fun) mean?

Custom function
Lambda function
Each value from df["Value"] is passed into fun() one by one.

### grouped_sum_based_on_region = df.groupby(['Product', 'Region'])["Value"].sum()
This line:

Groups your DataFrame by two columns: Product and Region.

For each unique combination of Product and Region, it sums up the Value column.

Returns a multi-index Series (hierarchical index for Product and Region).

```python
grouped_sum_based_on_region=df.groupby(['Product',"Region"])["Value"].sum()
grouped_sum_based_on_region
```

```python
grouped_mean_based_on_region=df.groupby(['Product',"Region"])["Value"].mean()
grouped_mean_based_on_region
```

###  agg(["mean", "sum"]) in Pandas
The .agg() function (short for aggregate) allows you to apply multiple functions (like mean, sum, min, max) at the same time on a DataFrame or Series.

```python
###  aggregate multiple function
grouped_agg=df.groupby(['Region','Product'])["Value"].agg(["mean","sum","count"])
grouped_agg
```

```python
### merging and joining dataframe

df1=pd.DataFrame({"key":['A','B','C','F'],"Value1":[12,54,76,29]})
df2=pd.DataFrame({"key":['A','B','E','D'],"Value1":[32,33,91,87]})
```

### pd.merge() combines two DataFrames (df1 and df2) by matching rows based on a common column or index.
It’s like a relational database join.

df1, df2	The two DataFrames you want to combine
on=" "	The column to merge on (must exist in both DataFrames)
how=" "	Type of join: inner, outer, left, or right

```python
## merge two datafream based on key
pd.merge(df1,df2,on="key",how="inner")  ##  Only keeps rows with matching ID in both DataFrames.
```

```python
pd.merge(df1,df2,on="key",how="outer")  ##  Keeps all rows from both DataFrames, fills missing values with NaN.
```

```python
pd.merge(df1,df2,on="key",how="left")  ## Keeps all rows from df1 (left DataFrame), matches from df2.
```

```python
pd.merge(df1,df2,on="key",how="right")  ##  Keeps all rows from df2 (right DataFrame), matches from df1.
```
