## Parametric Density Estimation

## ## normal() -> generates random numbers from a normal (Gaussian) distribution, which is the famous "bell curve."
loc=50 → mean (center) of the distribution is 50

scale=5 → standard deviation is 5 (how spread out the values are)

size=1000 → number of values to generate

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import normal

sample=normal(loc=50, scale=5,size=1000)
sample
```

```python
print(sample.mean(),sample.std())
```

```python
### plot a hsitrogram

plt.hist(sample,bins=10)   ## -> assume as a normal distribution
```

```python
sample_mean=sample.mean()
sample_std=sample.std()
```

Some key things scipy.stats can do: 

1️⃣ Probability Distributions

You can work with a variety of distributions:

Normal distribution → scipy.stats.norm

Binomial distribution → scipy.stats.binom

Poisson distribution → scipy.stats.poisson
and over 80 others.

```python
from scipy.stats import norm

dist=norm(sample_mean,sample_std) ## “Create a normal distribution with mean = sample_mean and standard deviation = sample_std.”
```

```python
values=np.linspace(sample.min(),sample.max(),100) ## randomly takes 100 elemets 

probability_densitys=[dist.pdf(val) for val in values]

## dist.pdf(50)       # probability density at x = 50
```

```python

plt.plot(values,probability_densitys)
```

```python
plt.hist(sample,bins=10,density=True)
```

## KDE

```python
sample1=normal(loc=10, scale=5,size=300)
sample2=normal(loc=30,scale=5,size=700)

sample=np.hstack((sample1,sample2))
sample
```

```python
plt.hist(sample,bins=50)
```

```python
from sklearn.neighbors import KernelDensity

model=KernelDensity(bandwidth=3,kernel="gaussian")

sample=sample.reshape((len(sample),1))

model.fit(sample)
```

```python
values=np.linspace(sample.min(),sample.max(),100)
values=values.reshape((len(values),1))
```

```python
probability_densitys=model.score_samples(values)
probability_densitys=np.exp(probability_densitys)
```

```python
plt.hist(sample,bins=50,density=True)
plt.plot(values[:],probability_densitys)
plt.show()
```
